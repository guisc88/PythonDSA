### Condicionais e Operadores.

```python
# Condicional if (se)... else (senão)
print(5>2) # booleano tradicional
print(5<2) # booleano tradicional
if 5 > 2:
    print('A sentença é verdadeira!')
else:
    print('A sentença não é verdadeira!')
    
if 5 < 2:
    print('A sentença é verdadeira!')
else:
    print('A sentença não é verdadeira!')
print()

# Condicional if (se)... else (senão) com variável
dia = 'Terça'
if dia == 'segunda':
    print('Hoje fará sol!')
else:
    print('Hoje vai chover!')
print()

# Podemos usar o operador elif para validar mais uma condição
dia = 'Terça'
if dia == 'segunda':
    print('Hoje fará sol!')
elif dia == 'terça':
    print('Hoje vai chover!')
else:
    print('Sem previsão do tempo para o dia selecionado')
print()

# Operadores relacionais em Pyhton
if 4 >= 4:
    print('Essa expressão é verdadeira')
else:
    print('Essa expressão não é verdadeira')

if 4 == 5:
    print('Essa expressão é verdadeira')
else:
    print('Essa expressão não é verdadeira')  
    
if 4 - 3 == 1:
    print('Essa expressão é verdadeira')
else:
    print('Essa expressão não é verdadeira') 
    
if 4 ** 2 == 15:
    print('Essa expressão é verdadeira')
else:
    print('Essa expressão não é verdadeira')     

a = 'Carro'
b = 'Vermelho'
if a + ' ' + b == 'Carro Vermelho':
    print('Essa expressão é verdadeira')
else:
    print('Essa expressão não é verdadeira')  
    
a = 'Carro '
b = 'Vermelho'
if a + b == 'Carro Vermelho':
    print('Essa expressão é verdadeira')
else:
    print('Essa expressão não é verdadeira')
print()    

# Condicionais aninhados
idade = 18
nome = 'Bob'
if idade > 13:
    if nome == 'Bob':
        print('Ok Bob, você está autorizado a entrar!')
    else:
        print('Desculpe, mas você não pode entrar!')
print()

# Operadores Lógicos
idade = 13
nome = 'Luis'
if nome == 'Luis' and idade >= 13:
    print('Ok Luis, você está autorizado a entrar!')
else:
    print('Desculpe, mas você não pode entrar!')
    
idade = 15
nome = 'Maria'
if (nome == 'Luis') or (idade >= 12):
    print('Ok, você está autorizado(a) a entrar!')
else:
    print('Desculpe, mas você não pode entrar!')  
    
cor = 'Branco'
raca = 'Poodle'
if not (raca == 'Poodle') and (cor == 'Branco'):
    print('Você não é um Poodle branco')
else:
    print('Você é um Poodle branco')
    
carro = 'branco'
valor = 100000
cidade = 'São Paulo'
if (carro != 'branco' and cidade != 'São Paulo') or valor <= 50000:
    print('Não podemos disponibilizar esse veículo no momento!')
else:
    print('Podemos encomendar o veículo!')

print()
```
